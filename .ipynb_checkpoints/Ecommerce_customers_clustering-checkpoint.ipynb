# E-commerce platform customers segmentation(clustering)

```python
import pandas as pd
```

```python
data=pd.read_csv("smartcart_customers.csv")
```

#### Our data 

```python
data.head()
```

#### Data cleaning

```python
data.isnull().sum()
```

```python
data['Income']=data['Income'].fillna(data['Income'].median())
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

### Feature engineering

```python
data['Age']=2026-data['Year_Birth']
```

```python
data['Total_Chlid']=data['Kidhome']+data['Teenhome']
```

```python
data['All_Spending']=data['MntWines']+data['MntFruits']+data['MntMeatProducts']+data['MntFishProducts']+data['MntGoldProds']+data['MntSweetProducts']
```

```python
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'],dayfirst=True)
recent_date=data['Dt_Customer'].max()
data['Total_join_days']=(recent_date-data['Dt_Customer']).dt.days
```

```python
data['Education'].value_counts()
```

```python
data['Education']=data['Education'].replace({'PhD':'Post_Graduation','Master':'Post_Graduation','2n Cycle':'Under_Graduate','Basic':'Under_Graduate'})
```

```python
data['Marital_Status'].value_counts()
```

```python
data['Living_behaviour']=data['Marital_Status'].replace({'Married':'Together','Single':'Alone','Divorced':'Alone','Widow':'Alone','Absurd':'Alone','YOLO':'Alone'})
```

```python
# Now dropping the unneccesary columns
fresh_data=data.drop(columns=['ID','Year_Birth','Marital_Status','Kidhome','Teenhome','Dt_Customer','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds'])
```

```python
# now we do feature encoding
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
New_encoded_data=ohe.fit_transform(fresh_data[['Education','Living_behaviour']])
# new dataframe making because of encoded
encoded_data=pd.DataFrame(New_encoded_data.toarray(),columns=ohe.get_feature_names_out(),index=fresh_data.index)
perfect_data=pd.concat([encoded_data,fresh_data],axis=1)
perfect_data=perfect_data.drop(columns=['Education','Living_behaviour'])
```

```python
perfect_data.info()
```

### Outlier detection and deletion

```python
# we detect it using plot
import seaborn as sns
sns.pairplot(perfect_data[['Income','Recency','Age','Total_Chlid','Total_join_days','All_Spending']])
```

```python
perfect_data=perfect_data[perfect_data['Age']<90]
perfect_data=perfect_data[perfect_data['Income']<600000]
```

```python
sns.pairplot(perfect_data[['Income','Recency','Age','Total_Chlid','Total_join_days','All_Spending']])
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(perfect_data)
```

```python
# now the all data preprocessing is done
```

### PCA

```python
# 2D
from sklearn.decomposition import PCA
pca=PCA(n_components=2,random_state=42)
pca_data2d=pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
sns.scatterplot(x=pca_data2d[:,0],y=pca_data2d[:,1])
```

```python
# 3D
from sklearn.decomposition import PCA
pca=PCA(n_components=3,random_state=42)
pca_data3d=pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(5,8))
fig.add_subplot(111,projection='3d')
plt.scatter(pca_data3d[:,0],pca_data3d[:,1],pca_data3d[:,2])
```

# Clustering using-KMeans and Agglomerative Algorithms

```python
# first find k value 
# 1 wcss(elbow method)
import warnings
warnings.filterwarnings('ignore') # just for ignoring warning
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i,random_state=42)
    cluster=kmean.fit_predict(X_scaled)
    wcss.append(kmean.inertia_)
```

```python
# k vs wcss plot
sns.lineplot(x=range(1,11),y=wcss,marker='>')
```

```python
# we can also use Kneed
from kneed import KneeLocator
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
```

```python
# now we got an k=4 total cluster=4
# so we use it on both algorithms
```

### KMeans clustering

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4,random_state=42)
clusters=km.fit_predict(pca_data3d)
```

```python
# visualizing clusters
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(5,8))
fig.add_subplot(111,projection='3d')
plt.scatter(pca_data3d[:,0],pca_data3d[:,1],pca_data3d[:,2],c=clusters)
```

```python
# we can seen the four clusters in plot
```

### AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
agg=AgglomerativeClustering(n_clusters=4,linkage='ward')
clusters=agg.fit_predict(pca_data3d)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,8))
fig.add_subplot(111,projection='3d')
plt.scatter(pca_data3d[:,0],pca_data3d[:,1],pca_data3d[:,2],c=clusters)
```

```python
# The agglomerative gives best clustering so we can say for this data the agglomerative is best
```

```python
# now add the cluster in our datasets for best understanding and finding the relation and customers type and behaviour
perfect_data['clusters']=clusters
```

```python
perfect_data.head()
```

## Capturing the meaninful info and differntiate the customers(Classification and Insights)
